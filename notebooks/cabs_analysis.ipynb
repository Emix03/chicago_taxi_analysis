### Evaluacion de hipotesis
## Hipotesisa evaluar '""La duración promedio de los viajes desde el Loop hacia el Aeropuerto Internacional O'Hare cambia los sábados en condiciones climáticas malas."

# H0 (Hipotesis nula) : 'La duración media de los viajes en sábados con clima "Bad" es igual a la de sábados con clima "Good"..'
# H1 (Hipotesis alternativa): La duración media de los viajes en sábados con clima "Bad" es diferente a la de sábados con clima "Good".

Como no conocemos de antemano si los viajes duran más o menos con mal clima, usamos prueba bilateral.

```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
df['start_ts'] = pd.to_datetime(df['start_ts']) #convert timestamps and extract by name 
df['day'] = df['start_ts'].dt.day_name()
```

```python
# Filetering only saturday days
saturday = df[df['day'] == 'Saturday']
```

```python
#Split data by weather condition
good_weather = saturday[saturday['weather_conditions'] == 'Good']['duration_seconds']
bad_weather  = saturday[saturday['weather_conditions'] == 'Bad']['duration_seconds']
```

```python
# 5. Print basic information
print("Trips with GOOD weather:", len(good_weather))
print("Trips with BAD weather:", len(bad_weather))
print("Mean duration GOOD:", good_weather.mean())
print("Mean duration BAD :", bad_weather.mean())
```

```python
#hypotesis test
stat, p =ttest_ind(good_weather,bad_weather,equal_var=False)
```

```python


print("\nT-statistic:", stat)
print("P-value:", p)

```

```python
alpha = 0.05  # significance level

if p < alpha:
    print("\n📌 RESULT: Reject H0 → Weather *does* affect trip duration on Saturdays.")
else:
    print("\n📌 RESULT: Fail to reject H0 → No evidence of change in trip duration.")
```

<div class="alert alert-block alert-success">

Bien hecho con el test de hipótesis, fue realizado de forma correcta. 

Es importante mencionar que el t-test asume que las dos muestras tienen varianzas iguales. Para verificar esa igualdad de varianzas podemos usar el **test de Levene**. Este nos dice si la variabilidad entre los grupos es estadísticamente diferente. En corto, el flujo sería así

1. Usamos `levene()` para verificar si las varianzas son iguales.
2. Dependiendo del resultado, usamos `ttest_ind()` con el argumento `equal_var`:
   - `equal_var=True` si Levene dice que las varianzas son iguales (p > 0.05).
   - `equal_var=False` si Levene dice que las varianzas son diferentes (p < 0.05).





</div>

## Conclusión — Prueba de hipótesis
Se evaluó la hipótesis: "La duración promedio de los viajes desde el Loop hacia el Aeropuerto Internacional O'Hare cambia los sábados bajo condiciones climáticas malas." Con un nivel de significancia α = 0.05, se compararon dos grupos de viajes realizados en sábado:
GOOD y BAD, La prueba estadística realizada fue un t-test de Welch, obteniendo:

t = -7.18

p-value = 6.73 × 10⁻¹²

Dado que el p-value es muchísimo menor que α, existe evidencia estadística suficiente para rechazar la hipótesis nula (H0) y aceptar la alternativa (H1).
## Interpretación
Los resultados indican que el clima sí influye significativamente en la duración de los viajes los sábados. En condiciones climáticas Bad, los viajes duran más en promedio, con una diferencia aproximada de 7 minutos adicionales frente a condiciones Good.
